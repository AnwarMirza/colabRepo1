# Demo Python Script

```python
#@title { vertical-output: true}
import numpy as np
import matplotlib.pyplot as plt
# Generate time axis (time in seconds - from 0 to 60 seconds in steps of 0.1)
t = np.arange(0.0, 60.0, 0.1)
# Frequency of signals in Hertz (1Hz = 1 /sec)
f = 1/60# Generate the signals
x = np.sin(2*np.pi*f*t)
y = np.cos(2*np.pi*f*t)
# Plot the signals
plt.plot(t, x, '-b', label='$x=sin(2\pi\,f\,t)$')
plt.plot(t, y, '-r', label='$y=cos(2\pi\,f\,t)$')
plt.legend(loc="lower left")
plt.xlabel('$t$')
plt.ylabel('Signal')
plt.grid(which='major', color='b', linestyle='-')
plt.grid(which='minor', color='g', linestyle='--')
plt.minorticks_on()
plt.show()
```
